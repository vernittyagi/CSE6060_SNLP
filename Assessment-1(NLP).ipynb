```python
#Importing the nltk package
import nltk
```

```python
#downloading the nltk data!!
nltk.download()
```

```python
#Importing the inaugural corpus
from nltk.corpus import inaugural
```

```python
#Taking a look at id's of the files in the corpus
inaugural.fileids()
```

```python
#Extracting out the text from the corpus
r_txt = inaugural.raw("2013-Obama.txt")
```

```python
print(r_txt)
```

```python
#Tokenizing the words!!
from nltk.tokenize import word_tokenize
print(word_tokenize(r_txt))
```

```python
#Extracting the POS_Tagged sentences from a corpus
brown_sents = nltk.corpus.brown.tagged_sents(categories="news")
for sent in brown_sents:
    for word, tag in sent:
        if (word == 'to' and tag == "IN"):
            print(sent)
```

```python
#Importing the stemmer and performing the stemming on those words!!
from nltk.stem import PorterStemmer
e_words= word_tokenize(r_txt)
ps =PorterStemmer()
for w in e_words:
    rootWord=ps.stem(w)
    print(rootWord)
```
