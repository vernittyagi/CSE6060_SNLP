```python
import nltk
```

```python
sent = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

```python
sen1 = "I observed that X had a keen interest and was fully involved in the course when I saw his performances during the Lab sessions, where he would be able to grasp new concepts such as query formation and joins."
```

```python
grammer = "NP: {<DT>?<JJ>*<NN>}"
```

```python
cp = nltk.RegexpParser(grammer)
```

```python
result = cp.parse(sent)
```

```python
print(result)
```

```python
result.draw()
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
result_lor = cp.parse(nltk.pos_tag(sen1))
```

```python
print(result_lor)
```

```python
result_lor.draw()
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
chunk = nltk.ne_chunk(nltk.pos_tag(sen1))
```

```python
print(chunk)
```
